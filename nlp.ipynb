```python
# imports
import nltk
from nltk.tokenize import sent_tokenize,word_tokenize,PunktSentenceTokenizer
from nltk.corpus import stopwords,state_union
from nltk.stem import PorterStemmer,LancasterStemmer,WordNetLemmatizer
```

```python
example_text="Hello Mr. AK how are you doing today? Javascript is beautiful and python is awesome."

# tokenize by word
print(word_tokenize(example_text))
```

```python
example_text="Hello Mr. AK how are you doing today? Javascript is beautiful and python is awesome."
# tokenize by sentence
print(sent_tokenize(example_text))
```

```python
# filter out stop words
stop_words=set(stopwords.words("english"))
words=word_tokenize(example_text)
filtered_sentence=[w for w in words if w not in stop_words]
print(filtered_sentence)
```

```python
#Library to form wordcloud
text="Hello Mr. AK how are you doing today? Javascript is beautiful and python is awesome."
from wordcloud import WordCloud

#library to plot the wordcloud

import matplotlib.pyplot as plt

#generatin

wordcloud = WordCloud().generate(text)

#plot the wordcloud

plt.figure(figsize =(12,12))

plt.imshow(wordcloud)

#to remove thw axis value

plt.axis("off")

plt.show()
```

```python
# stemming
stemmer=PorterStemmer()
words=['give','given','gives','giving']
stemmed=[stemmer.stem(w) for w in words]
print(stemmed)
stemmer=LancasterStemmer()
stemmed=[stemmer.stem(w) for w in words]
print(stemmed)
```

```python
#lemmatizing
lem = WordNetLemmatizer()

print(lem.lemmatize("studying", pos ="v"))

print(lem.lemmatize("studying", pos ="n"))

print(lem.lemmatize("studying", pos ="a"))

print(lem.lemmatize("studying", pos ="r"))
```

```python
# part of speech tagging
train_text=state_union.raw('2005-GWBush.txt')
text=state_union.raw('2006-GWBush.txt')
custom_sent_tokenizer=PunktSentenceTokenizer(train_text)
tokenized=custom_sent_tokenizer.tokenize(text)

def process_content():
    try:
        for i in tokenized:
            words=word_tokenize(i)
            tagged=nltk.pos_tag(words)
            print(tagged)
    except Exception as e:
        print(str(e))

process_content()
```
